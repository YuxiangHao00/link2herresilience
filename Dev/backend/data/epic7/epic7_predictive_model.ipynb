### Dependent variables
For the determination of the risk related to following ending with:
- "emctd": Did you continue to use $substance$ even though you thought this was causing you to have problems with your emotions, nerves, or mental health? 
- "phlpb": During the past 12 months, did you have any physical health problems that were probably caused or made worse by your use of $substance$?
- "phctd" Did you continue to use $substance$ even though you thought this was causing you to have physical problems?, 
- "lsact": During the past 12 months, did using $substance$ cause you to give up or spend less time doing these types of important activities such as working, going to school, taking care of children, doing fun things such as hobbies and sports,
and spending time with friends and family?, 
- "serpb": During the past 12 months, did using $substance$ cause you to have serious problems like this either at home, work, or school?, 
- "pdang": During the past 12 months, did you regularly use $substance$ and then do something where using $substance$ might have
put you in physical danger?,                
- "lawtr": During the past 12 months, did using $substance$ cause you to do things that repeatedly got you in trouble with the law?, 
- "frfpb": During the past 12 months, did you have any problems with family or friends that were probably caused by your use of $substance$?, 
- "fmctd": Did you continue to use $substance$ even though you thought this caused problems with family or friends?

### Data Pre-processing w/ pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# pre-processor: ColumnTransformer
# standardise numerical attributes, 1-hot encoding for categorical
pre_processor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), ["cig30use", "iralcfy", "irmjfy"]),
    ('cat', OneHotEncoder(drop='first'), set(df_subset.columns) - {"cig30use", "iralcfy", "irmjfy"})
]
)
```

### Predictive modelling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm

from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
# Parameter grid for GridSearchCV
model_grid = {
    'random_forest': {
        'model':RandomForestClassifier(random_state=15, n_jobs=3, n_estimators=500),
        'params': {
            'estimator__max_depth': [11, 12, 13, 14],
            'estimator__criterion':['gini', 'entropy']
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(random_state=15, n_jobs=3),
        'params': {
            'estimator__C': [0.085, 0.09, 0.092],
            'estimator__solver':['lbfgs', 'liblinear'],
        }
    },
    'svm': {
        'model': svm.LinearSVC(random_state=15, max_iter=10000),
        'params': {
            'estimator__C':[0.52, 0.55, 0.6, 0.65]
        }
    }
}
```

```python
# list of model scores
model_scores = []

for model_name, model_params in model_grid.items():
    pipe = Pipeline(steps=[
        ('preprocessor', pre_processor),
        ('estimator', model_params['model'])
    ])

    model = GridSearchCV(estimator=pipe, 
                         param_grid=model_params['params'], 
                         cv=5, return_train_score=False, refit=True)
    model.fit(df_X.fillna(0, inplace=True), df_cause_effect.fillna(0, inplace=True))
    model_scores.append({
        'model': model_name,
        'best_score:': model.best_score_,
        'best_params': model.best_params_
    })
```
