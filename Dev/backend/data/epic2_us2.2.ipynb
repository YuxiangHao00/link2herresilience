```python
import os
import pandas as pd
import re
```

```python
dir_path = os.getcwd() + "/epic2/2.2/"
list_pollen_in_state_json = os.listdir(dir_path+'tern_data/')

dict_index_pollen_in_state_json = {}
for i, file in enumerate(list_pollen_in_state_json):
    if file[-5:] == '.json' and file != 'pollen_grains_in_state.json':
        curr_state = re.search(r'\b([A-Z]+)_', file).group(1)
        dict_index_pollen_in_state_json[curr_state.lower()] = i
```

```python
set_aus_states = set(dict_index_pollen_in_state_json.keys())
set_aus_states
```

### VIC

```python
import json
```

```python
dict_pollen_grains_in_state = {}

def get_pollen_grains_in_state(state):
    path_json = dir_path + "tern_data/"+ list_pollen_in_state_json[dict_index_pollen_in_state_json[state]]
    # dict_df_pollen_count_in_state['vic'] = pd.read_json(path_json)
    with open(path_json, 'r') as file:
        state_pollen_data = json.load(file)

    pollen_grains_in_state = set()
    
    for i, features in enumerate(state_pollen_data["features"]):
        str_pollen_grains_keys = " ".join(features["properties"].keys())
        # print(str_pollen_grains_keys)
        
        pollen_grains_in_state = pollen_grains_in_state | set(re.findall(r' ([A-Za-z]*)_pollen_week', str_pollen_grains_keys))

    try:
        pollen_grains_in_state.remove('other')
        pollen_grains_in_state.remove('total')
    except:
        pass
        
    return {'location_name': state_pollen_data["features"][0]["properties"]["location"],
            'pollen_grains': list(pollen_grains_in_state),
            'coordinates': state_pollen_data["features"][0]["geometry"]["coordinates"][0]}
```

```python
dict_pollen_grains_in_state['vic'] = get_pollen_grains_in_state('vic')
```

### pollen counts

```python
def get_pollens_count_in_state(state):
    path_json = dir_path + "tern_data/" + list_pollen_in_state_json[dict_index_pollen_in_state_json[state]]
    # dict_df_pollen_count_in_state['vic'] = pd.read_json(path_json)
    with open(path_json, 'r') as file:
        state_pollen_data = json.load(file)

    dict_pollens_count_in_state = {}
    
    for i, feature in enumerate(state_pollen_data["features"]):
        for pollen in dict_pollen_grains_in_state[state]["pollen_grains"]:
            if pollen not in dict_pollens_count_in_state.keys():
                dict_pollens_count_in_state[pollen] = {'year': [], 'week_number': [], 'count': []}
            try:
                dict_pollens_count_in_state[pollen]['year'].append(feature["properties"]["first_date_of_week"][:4])
                dict_pollens_count_in_state[pollen]['week_number'].append(feature["properties"]["week_number"])
                dict_pollens_count_in_state[pollen]['count'].append(feature["properties"][pollen+"_pollen_week_count_grains_metre_cubed"])
            except:
                continue
            
        
    return dict_pollens_count_in_state


```

```python
dict_pollens_count_in_aus = {}
for state in set_aus_states:
    dict_pollens_count_in_aus[state] = get_pollens_count_in_state(state)

    with open(dir_path + state + "_pollens_count.json", 'w') as file:
        json.dump(dict_pollens_count_in_aus[state], file)

dict_pollens_count_in_aus['vic']
```
